## Feature detection с испльзованием SIFT детектора

```python
import os, sys
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import matplotlib.image as mpimg
```

```python
print('OpenCV version: {}'.format(cv2.__version__))
```

```python
abs_path = os.path.dirname(os.path.realpath('__file__'))
img_name = "\St-Pauls-Cathedral"
img_type = ".jpg"
folder_name = '\demo-data'
img_path = "{}{}{}".format(abs_path, folder_name, img_name + img_type)
```

```python
image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
```

```python
original = mpimg.imread(img_path)
plt.imshow(original)
```

```python
print('size: {} shape: {}'.format(image.size, image.shape))
```

```python
plt.imshow(image)
```

```python
sift = cv2.xfeatures2d.SIFT_create()
keypoints = sift.detect(image, None)
```

```python
result = cv2.drawKeypoints(image, keypoints, image, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
output_path = '{}{}-{}'.format(abs_path + folder_name, img_name, "result" + img_type)
cv2.imwrite(output_path, result)

plt.imshow(result)
plt.show()
```

```python
keypoints, descriptor = sift.compute(image, keypoints)
print('descriptor: {}'.format(len(descriptor)))
```

## Используем SURF и ORB

```python
surf = cv2.xfeatures2d.SURF_create()
orb = cv2.ORB_create(nfeatures=1500)
```

```python
keypoints_surf, descriptor_surf = surf.detectAndCompute(image, None)
keypoints_orb, descriptors_surf = orb.detectAndCompute(image, None)
```

```python
result = cv2.drawKeypoints(image, keypoints_surf, image)
output_path = '{}{}-{}'.format(abs_path + folder_name, img_name, "result_surf" + img_type)
cv2.imwrite(output_path, result)

plt.imshow(result)
plt.show()
```

```python
result = cv2.drawKeypoints(image, keypoints_orb, image)
output_path = '{}{}-{}'.format(abs_path + folder_name, img_name, "result_orb" + img_type)
cv2.imwrite(output_path, result)

plt.imshow(result)
plt.show()
```
